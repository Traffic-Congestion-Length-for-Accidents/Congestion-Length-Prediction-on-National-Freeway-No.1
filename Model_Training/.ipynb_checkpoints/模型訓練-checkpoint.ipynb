```python
def score_result(y_test, pred):
    TP = np.sum((y_test == 1) & (pred == 1))
    FN = np.sum((y_test == 1) & (pred == 0))
    TN = np.sum((y_test == 0) & (pred == 0))
    FP = np.sum((y_test == 0) & (pred == 1))

    Recall = TP/(TP+FN)
    Accuracy = (TP+TN)/(len(y_test))
    Precision = (TP)/(TP+FP)
    F1_score = (2*Precision*Recall)/(Precision+Recall)
    
    print(f"Recall: {Recall} \n Accuracy: {Accuracy} \n Precision: {Precision} \n F1_score: {F1_score}")
```

**<font size = "4">以Accuracy及F1-score指標選用CatBoost模型，並針對其中<font color = "red">樹的深度(depth)</font>以及<font color = "red">樹的數量(iterations)</font>調整參數</font>**

## 尋找最佳的初始樣本的權重組合
<font size = "3.5">根據右圖的混淆矩陣，本研究認為<font color = "red">有回堵卻預測成不回堵(即Fasle Negative)引起的道路災情更嚴重</font>，也較容易引起民怨。</font>\
<img align="right" src="混淆矩陣.png"/>
<font size = "3.5"> 參考兩個指標分別為</font>


<font size = "3.5">1.                                 
        Recall = $\frac{PT}{PT+FN}$</font>

<font size = "3.5">2.\
        F-score = $\frac{(1+\beta^{2})*Precision*Recall}{(\beta^{2}*Precision)+Recall}$</font>

<font size = "3.5">F2-score即$\beta$為2時，權重會偏向Recall一些。而要降低FN的發生，就期望Recall值能提高，故參考此二指標調整初始樣本的權重組合</font>


```python
weight_1 = [i for i in np.arange(0.5, 1, 0.01)]
recall_value = []
precision_record = []
accu_record = []
F1_record = []
F2_record = []

for i in weight_1:
    
    weight_2 = 1-i
    model = CatBoostClassifier(iterations = 950, depth = 4,\
                               cat_features = cate,verbose =False, class_weights = {0: weight_2 , 1 : i})
    model.fit(x_train, y_train)
    pred = model.predict(x_test)
    
    TP = np.sum((y_test == 1) & (pred == 1))
    FN = np.sum((y_test == 1) & (pred == 0))
    TN = np.sum((y_test == 0) & (pred == 0))
    FP = np.sum((y_test == 0) & (pred == 1))

    Recall = TP/(TP+FN)
    Accuracy = (TP+TN)/(len(y_test))
    Precision = (TP)/(TP+FP)
    F1_score = (2*Precision*Recall)/(Precision+Recall)
    F2_score = (5*Precision*Recall)/(4*Precision+Recall)
    
    recall_value.append(Recall)
    precision_record.append(Precision)
    accu_record.append(Accuracy)
    F1_record.append(F1_score)
    F2_record.append(F2_score)
```

```python
plt.figure(figsize = [10,10])
plt.plot(weight_1, recall_value, "-.r")

plt.plot(weight_1, F2_record, '-.d')
plt.xlabel("有回堵資料占樣本比例(weight)",  fontproperties = winfont02, size = 15)
plt.legend(["Recall", "F2-score"])

plt.axvline(weight_1[np.argmax(F2_record)], color = "orange", ls = "--", alpha = 0.3)
plt.text(weight_1[np.argmax(F2_record)]+0.008, max(F2_record)+0.01,\
         f"Max F2-score: {round(max(F2_record),4)}\nRecall: {round(recall_value[np.argmax(F2_record)],4)}\nCongestion weight in sample : {round(weight_1[np.argmax(F2_record)],2)}")


plt.title("CatBoost不同樣本權重下模型表現", fontproperties = winfont02, size = 18)

plt.show()
```

<font size = "4">最後本研究對初始樣本比例選擇**有回堵0.87**，而**無回堵0.13**。</font>

### 各階段模型訓練集以及最後測試集設置

### 第一階段模型 :
- 訓練集 : 依變數 = x_train_first , 自變數 = y_train_first
### 第二階段模型 :
- 訓練集 : 依變數 = x_train_second , 自變數 = y_train_second

### 測試集 :  
- 依變數 = x_test , 自變數 = y_test

```python
x_train_first = ori_data.drop(columns = ["Congestion", "CongestionMileage"])
y_train_first = ori_data.Congestion


x_train_second = ori_data.query("Congestion == 1")
x_train_second.index = [i for i  in range(0, x_train_second.shape[0])]

y_train_second = np.log(x_train_second.CongestionMileage)

x_train_second = x_train_second.drop(columns = ["Congestion", "CongestionMileage"])




x_test = test_data.drop(columns = ["Congestion", "CongestionMileage"])
y_test = test_data.CongestionMileage
```

### 分別建置各階段最終模型

```python
# 第一階段模型訓練 - CatBoost
cate = np.array(['Direction', 'DateCategory', 'TimeCategory'])
model_stage1 = CatBoostClassifier(iterations = 950, depth = 4,cat_features = cate,verbose =False,\
                           class_weights = {0: 0.13 , 1 : 0.87})

model_stage1.fit(x_train_first, y_train_first)


scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(x_train_second)


#這裡以下要接尾加他們的 因為要串起整個模型流程

# 第二階段模型訓練 - MLP
model_stage2 = Sequential([
                            Dense(64, input_dim=X_train_scaled.shape[1], activation='relu'),
                            Dropout(0.3),
                            Dense(32, activation='relu'),
                            Dropout(0.3),
                            Dense(32, activation='relu'),
                            Dropout(0.2),
                            Dense(1, activation='linear')
                        ])

model_stage2.compile(optimizer='adam', loss='mse')

model_stage2.fit(X_train_scaled, y_train_second, epochs=100, batch_size=32, verbose=1)
```

### 根據流程圖計算誤差及低估比例


```python
#第一階段預測
pred = model_stage1.predict(x_test)

#紀錄預測之回賭里程
result = []

for index, value in enumerate(pred):
    if value == 0:      #若第一階段預測為0則回堵里程為0
        result.append(0)
    else:  #若第一階段預測為1則再由第二階段模型預測回堵里程
        temp_dd = pd.DataFrame(x_test.loc[index,:]).T
        X_test_scaled = scaler.transform(temp_dd)
        result.append(np.exp(model_stage2.predict(X_test_scaled))[0][0])
        
under_rate = ((result < y_test).sum()/len(y_test))*100   #低估比例
mse = mean_squared_error(y_test, result)    #MSE
```
